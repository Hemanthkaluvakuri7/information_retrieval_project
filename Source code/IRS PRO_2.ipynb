```python
import nltk
import os
import re
import sys


nltk.download('punkt')
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
#Getting the folder in which text files are present in my system. You can try replacing with your path.
path='/Users/hemanthkaluvakuri/Downloads/test_data/'
files = os.listdir(path)
stemm = PorterStemmer()
count=0
fr=[]
sp=[]
dc=[]
dcc=[]
dco=0
dcid=[]
tst=0
#getting stop words from the path in my system. You can try replacing with your path.
stopw=open('/Users/hemanthkaluvakuri/Downloads/project1-2/stopwordlist.txt')

stw=[]
for x in stopw:
    stw.append(x.rstrip('\n').replace(" ", ""))


for fil in files:
    f  = open(path+fil, mode='r', encoding='utf-8')
    for x in f:
        if '<DOCNO>' in x:
            x=(x.split('<DOCNO>'))[1].split('</DOCNO>')[0]
            dco=dco+1
            dcid.append(x)
        if(x.find('<TEXT>') == 0 or x.find('</TEXT>') == 0):
            count=count+1
        if(count%2 == 1):
            words=x.split()
            for y in words:
                if(y!='<TEXT>'):
                    if not (re.findall(r'[0-9]+', y)):
                        if(re.findall(r'[`\-=~!@#$%^&*()_+\[\]{};\'\\:"|<,./<>?]', y)):
                            sp=[]
                            sp.append(re.split(r'[`\-=~!@#$%^&*()_+\[\]{};\'\\:"|<,./<>?]', y))
                            for s in sp:
                                fr.extend(s)
                                for r in s:
                                    dcc.append(dco)
                                
                        if not (re.findall(r'[`\-=~!@#$%^&*()_+\[\]{};\'\\:"|<,./<>?]', y)):
                            fr.append(y)
                            dcc.append(dco)

                            

fk=[]
dk=[]
for i in range(len(fr)):
    if fr[i].lower() not in stw:
        if(fr[i]!=''):
            fk.append(stemm.stem(fr[i].lower()))
            dk.append(dcc[i])

xyz=[]
for i in range(len(fk)):
    t=fk[i],dk[i]
    xyz.append(t)

#Printing the result into the path in my system. You can try replacing with your path.
with open('/Users/hemanthkaluvakuri/Downloads/Forward_Index2.txt', 'w') as f:
    te=[]
    docst=''
    h=0
    inte=[]
    print('DOCUMENT      WORDS', file=f)
    while(h<xyz[len(xyz)-1][1]):
        te2=[]
        for i in range(len(xyz)-1):
            if(xyz[i][1]==h+1):
                te.append(xyz[i][0])
                docst=docst+';'+xyz[i][0]
        te=docst.split(';')
        te=sorted(te)
        for t in te:
            if(t!=''):
                xxx=t+':'+str(te.count(t))
                te2.append(xxx)
        te2=list(dict.fromkeys(te2))
        docst=str(dcid[h])
        for x in te2:
            docst=docst+' ; '+x
        print(docst, file=f)
        inte.append(docst)
        print('--------------------------', file=f)              
        h=h+1;
        docst=''

#Printing the result into the path in my system. You can try replacing with your path.
with open('/Users/hemanthkaluvakuri/Downloads/Inverted_Index2.txt', 'w') as f:
    inte2=[]
    fk2=[]
    fk2=sorted(fk)
    fk2=list(dict.fromkeys(fk2))
    for i in fk2:
        sti=" "+i+':'
        docst=i
        for j in range(len(inte)):
            if sti in inte[j]:
                pattern = sti+"(.*?)\ ;"
                try:
                    substring = re.search(pattern, inte[j]).group(1)
                except:
                    inte[j]=inte[j]+' '
                    pattern = sti+"(.*?)\ "
                    substring=re.search(pattern, inte[j]).group(1)
                docst=docst+' ; '+dcid[j]+':'+substring
        inte2.append(docst)       
        print(docst, file=f)
        
inn=[]
word='Y'
while(word=='Y' or word =='y'):
    word=input("Enter the word:")
    if word in stw:
        print(word,'is a stop word')
    elif(word!=''):
        for i in inte2:
            inn=i.split()
            if word==inn[0]:
                print(i)
    word=input('Do you want to enter another word: Y/N or y/n  : ')


    


    
            


        
```
